# Task 1: Conceptual Understanding

### "Love" vs "love":
In NLP, these are treated as different tokens due to case sensitivity. "Love" (capitalized) might be seen as a proper noun or start of a sentence, while "love" is a standard verb/noun. Standardizing to lowercase ensures the model treats them as the same concept.

### Missing Stopword Removal:
If stopwords (common words like "the", "is", "at") aren't removed, they can dominate frequency counts and "drown out" the unique words that actually carry the sentence's meaning.

### Harmful Stopword Removal Scenarios:
Sentiment Analysis: In the phrase "not happy," removing the stopword "not" changes the meaning from negative to positive.
Search Queries: In a search for "To be or not to be," removing stopwords would leave almost no searchable terms.

### Stemming vs Lemmatization:
 Stemming is a "brute-force" approach that chops off word endings (e.g., "studies" becomes "studi"). Lemmatization uses a dictionary to find the root word based on its actual meaning (e.g., "better" becomes "good").

```python
import re
from collections import Counter
```

```python
def preprocess_text(text):
    if not text or not isinstance(text, str):
        return []

    text = text.lower() # Lowercase
    text = re.sub(r'https?://\S+|www\.\S+|\S+@\S+', '', text) # Remove URLs/Emails
    text = re.sub(r'(.)\1{2,}', r'\1', text) # Handle repetitions (sooo -> so)
    text = re.sub(r'\d+', '', text) # Remove numbers
    text = re.sub(r'[^a-z\s]', '', text) # Remove special chars/emojis
    text = " ".join(text.split()) # Remove extra spaces

    # Remove short tokens (<=2) but keep "no", "not"
    tokens = text.split()
    keep = {'no', 'not'}
    return [t for t in tokens if len(t) > 2 or t in keep]
```

```python
# Task 6: Build Full Pipeline


def full_pipeline(text_list):
    res = {"tokens": [], "clean_sentences": []}
    for text in text_list:
        # Task 7: Handling empty/null/emoji-only strings
        tokens = preprocess_text(text)
        res["tokens"].append(tokens)
        res["clean_sentences"].append(" ".join(tokens))
    return res


```
