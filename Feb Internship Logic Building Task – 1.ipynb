```python
marks = [45, 78, 90, 33, 60]

pass_students = []
fail_students = []

for mark in marks:
    if mark < 50:
        fail_students.append(mark)
    else:
        pass_students.append(mark)

print(f"count of pass students: {len(pass_students)}")
print(f"count of fail students: {len(fail_students)}")


```

```python
# 3. Simple Data Cleaner
# Real-World Application: Data preprocessing before analysis

names = [" Alice ", "bob", " CHARLIE "]

# Requirements: Remove extra spaces and convert to lowercase
cleaned_names = [name.strip().lower() for name in names]

# Print the cleaned list
print("Cleaned Names:", cleaned_names)
```

```python
# Message Length Analyzer 
messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length = len(msg)
    # Print the length of each message
    print(f"Message: '{msg}' | Length: {length}")
    
    # Flag messages longer than 10 characters
    if length > 10:
        print(f"  -> Flagged: '{msg}' is too long.")
```

```python
# Error Message Detector 

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Count the number of "ERROR" entries
error_count = logs.count("ERROR")

# Print the total error count
print(f"Log Analysis: Found {error_count} errors.")
```
